*  add Product in the PriceList

*  compute the total value of the PriceList (given by the sum of the prices of each product contained in the PriceList)

```python
# code here the class PriceList considering the functionalities mentioned before
class PriceList:
  def __init__(self, lstprc):
    self.lstprc = lstprc
  
  def add_product(self, product):
    self.lstprc.append(product)
  
  def find_price(self, product_ID):
    for x in self.lstprc:
      if x.get_ID() == product_ID:
        print("Il prezzo è: " + str(x.get_price()))

  def total(self):
    sum = 0
    for x in self.lstprc:
      sum += x.get_price()
    return sum
  
```

```python
# Use the module copy
import copy
old_list = [[1, 1, 1], [2, 2, 2], [3, 3, 3]]

shlist = copy.deepcopy(old_list)

print("Stampa delle liste prima delle modifiche")
print(shlist)
print(old_list)

shlist[1] = [2,2,'BB']

print("Stampa delle liste dopo modifica a shlist")
print(shlist)
print(old_list)

old_list[1] = [2,2,'CC']

print("Stampa delle liste dopo modifica a old_list")
print(shlist)
print(old_list)
```

Write a Python class Restaurant with attributes:
*   menu_items
*   booked_table
*   customer_orders

and methods:
*   add_item_to_menu(item, price)
*   book_tables(table)
*   customer_order(table, item)

In the main, perform the following tasks now:

    Now add items to the menu.
    Make table reservations.
    Take customer orders.
    Print the menu.
    Print table reservations.
    Print customer orders.

Note: Use dictionaries and lists to store the data.

```python
class Resturant:
    def __init__(self, menu_items, booked_table, customer_orders):
        self.menu_items = menu_items
        self.booked_table = booked_table
        self.customer_orders = customer_orders
    
    def add_item_to_menu(self, item, price):
        self.menu_items.update({item : price})
    
    def book_tables(self, table):
        if self.booked_table[table] == 0:
            self.booked_table[table] = 1
        else:
            print("Tavolo già occupato")

    def customer_order(self, table, item):
        if self.booked_table[table] == 0:
            print("Tavolo non prenotato, devi prima prenotarlo")
        else:
            self.customer_orders.update({table : item})

    def print_menu(self):
        print(self.menu_items)
    
    def print_table(self):
        for x in self.booked_table:
            if self.booked_table[x] == 0:
                print("Tavolo N° " + str(x) + " disponibile")
            else: 
                if self.booked_table[x] == 1:
                    print("Tavolo N° " + str(x) + " occupato")
    
    def print_order(self):
        for x in self.customer_orders:
            print("Il tavolo " + str(x) + " ha ordinato " + str(self.customer_orders[x]))
```

```python
myMenu = {
    "Spaghetti" : 10,
    "Ravioli" : 15,
    "Pollo" : 12,
    "Filetto" : 20
}

occupiedTable = {
    0 : 0,
    1 : 0,
    2 : 0,
    3 : 0,
    4 : 1,
    5 : 1,
}

order = {
    5 : ["Spaghetti", "Ravioli", "Pollo"],
    4 : ["Filetto"]
}

Cracco = Resturant(myMenu,occupiedTable,order)

Cracco.print_menu()
Cracco.add_item_to_menu("Pizza", 10)
Cracco.print_menu()

Cracco.print_table()

Cracco.book_tables(2)

Cracco.print_table()

Cracco.book_tables(5)

Cracco.print_order()

Cracco.customer_order(2,["Spaghetti","Pizza"])

Cracco.print_order()
```
